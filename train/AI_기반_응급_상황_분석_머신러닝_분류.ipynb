```python
import pandas as pd
import requests

# base URL과 파일 리스트 정의
base_url = "<data_url>"
train_files = [
    'TL_광주_구급.json', 'TL_광주_구조.json', 'TL_광주_기타.json', 'TL_광주_화재.json',
    'TL_서울_구급.json', 'TL_서울_구조.json', 'TL_서울_기타.json', 'TL_서울_화재.json',
    'TL_인천_구급.json', 'TL_인천_구조.json', 'TL_인천_기타.json', 'TL_인천_화재.json'
]

train_dfs = []

for file_name in train_files:
    # URL 인코딩 및 파일 불러오기
    url = base_url + quote(file_name)
    try:
        # JSON 파일 읽기
        df = pd.read_json(url)
        train_dfs.append(df)
        print(f"{file_name} 파일을 성공적으로 불러왔습니다.")
    except Exception as e:
        print(f"{file_name} 파일을 불러오는 중 오류 발생: {e}")

# 모든 데이터를 하나의 데이터프레임으로 합치기
train_data = pd.concat(train_dfs, ignore_index=True)
```

```python
train_data_subset = train_data[['disasterMedium','urgencyLevel','sentiment','triage']]
# train_data_subset을 복사하여 원본을 수정하지 않도록 처리
train_data_subset = train_data_subset.copy()

# triage 열에서 빈 문자열을 '무증상'으로 변경
train_data_subset.loc[train_data_subset['triage'].apply(lambda x: x == ''), 'triage'] = "무증상"

# 'None' 값을 '무증상'으로 변경 (NaN 처리)
train_data_subset.loc[train_data_subset['triage'].isna(), 'triage'] = "무증상"
train_data_subset
```

```python
train_data_subset.to_csv('train_data_subset.csv', index=False)
```

```python
base_url = "https://media.githubusercontent.com/media/1000century/ml/master/valid_label/"
valid_files = [file.replace('TL_', 'VL_') for file in train_files]
valid_dfs = []

for file_name in valid_files:
    # URL 인코딩 및 파일 불러오기
    url = base_url + quote(file_name)
    try:
        # JSON 파일 읽기
        df = pd.read_json(url)
        valid_dfs.append(df)
        print(f"{file_name} 파일을 성공적으로 불러왔습니다.")
    except Exception as e:
        print(f"{file_name} 파일을 불러오는 중 오류 발생: {e}")

# 모든 데이터를 하나의 데이터프레임으로 합치기
valid_data = pd.concat(valid_dfs, ignore_index=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

# 데이터 로드
train_data = pd.read_csv('train_data_subset.csv')
valid_data = pd.read_csv('valid_data_subset.csv')

# 각 열별로 LabelEncoder 인스턴스 생성
disasterMedium_encoder = LabelEncoder()
sentiment_encoder = LabelEncoder()
triage_encoder = LabelEncoder()
urgencyLevel_encoder = LabelEncoder()

# 훈련 데이터에 대해서 fit
train_data['disasterMedium'] = disasterMedium_encoder.fit_transform(train_data['disasterMedium'])
train_data['sentiment'] = sentiment_encoder.fit_transform(train_data['sentiment'])
train_data['triage'] = triage_encoder.fit_transform(train_data['triage'])
train_data['urgencyLevel'] = urgencyLevel_encoder.fit_transform(train_data['urgencyLevel'])

# 검증 데이터는 훈련 데이터에서 학습된 레이블로 변환
valid_data['disasterMedium'] = disasterMedium_encoder.transform(valid_data['disasterMedium'])
valid_data['sentiment'] = sentiment_encoder.transform(valid_data['sentiment'])
valid_data['triage'] = triage_encoder.transform(valid_data['triage'])
valid_data['urgencyLevel'] = urgencyLevel_encoder.transform(valid_data['urgencyLevel'])

# 특징과 타겟 변수 설정
X_train = train_data[['disasterMedium', 'sentiment', 'triage']]
y_train = train_data['urgencyLevel']

X_valid = valid_data[['disasterMedium', 'sentiment', 'triage']]
y_valid = valid_data['urgencyLevel']

# Random Forest 모델 설정 및 하이퍼파라미터 튜닝
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]  # 'auto' 대신 'sqrt' 사용
}

# GridSearchCV를 사용하여 하이퍼파라미터 튜닝
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid,
                           cv=5,
                           n_jobs=-1,
                           verbose=2)

# 모델 학습
grid_search.fit(X_train, y_train)

# 최적의 하이퍼파라미터 출력
print(f"Best parameters: {grid_search.best_params_}")

# 최적의 모델로 예측
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_valid)

# 모델 평가
accuracy = accuracy_score(y_valid, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_valid, y_pred))

# 교차 검증을 통해 모델 성능 확인
cv_scores = cross_val_score(best_model, X_train, y_train, cv=5)
print(f"Cross-validated accuracy scores: {cv_scores}")
print(f"Mean CV accuracy: {cv_scores.mean()}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

# 데이터 로드
train_data = pd.read_csv('train_data_subset.csv')
valid_data = pd.read_csv('valid_data_subset.csv')

# 각 열별로 LabelEncoder 인스턴스 생성
disasterMedium_encoder = LabelEncoder()
sentiment_encoder = LabelEncoder()
triage_encoder = LabelEncoder()
urgencyLevel_encoder = LabelEncoder()

# 훈련 데이터에 대해서 fit
train_data['disasterMedium'] = disasterMedium_encoder.fit_transform(train_data['disasterMedium'])
train_data['sentiment'] = sentiment_encoder.fit_transform(train_data['sentiment'])
train_data['triage'] = triage_encoder.fit_transform(train_data['triage'])
train_data['urgencyLevel'] = urgencyLevel_encoder.fit_transform(train_data['urgencyLevel'])

# 검증 데이터는 훈련 데이터에서 학습된 레이블로 변환
valid_data['disasterMedium'] = disasterMedium_encoder.transform(valid_data['disasterMedium'])
valid_data['sentiment'] = sentiment_encoder.transform(valid_data['sentiment'])
valid_data['triage'] = triage_encoder.transform(valid_data['triage'])
valid_data['urgencyLevel'] = urgencyLevel_encoder.transform(valid_data['urgencyLevel'])

# 특징과 타겟 변수 설정
X_train = train_data[['disasterMedium', 'sentiment', 'triage']]
y_train = train_data['urgencyLevel']

X_valid = valid_data[['disasterMedium', 'sentiment', 'triage']]
y_valid = valid_data['urgencyLevel']

# Random Forest 모델 설정 및 하이퍼파라미터 튜닝
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]  # 'auto' 대신 'sqrt' 사용
}

# RandomForestClassifier에서 class_weight='balanced'로 설정
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42, class_weight='balanced'),
                           param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# 모델 학습
grid_search.fit(X_train, y_train)

# 최적의 하이퍼파라미터 출력
print(f"Best parameters: {grid_search.best_params_}")

# 최적의 모델로 예측
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_valid)

# 모델 평가
accuracy = accuracy_score(y_valid, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_valid, y_pred))

# 교차 검증을 통해 모델 성능 확인
cv_scores = cross_val_score(best_model, X_train, y_train, cv=5)
print(f"Cross-validated accuracy scores: {cv_scores}")
print(f"Mean CV accuracy: {cv_scores.mean()}")
```
